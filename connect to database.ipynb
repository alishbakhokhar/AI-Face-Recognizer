```python
import mysql.connector

# ------------------- CONNECT -------------------
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="authorized_user"
)

mycursor = mydb.cursor()

# ------------------- FIX COLUMN NAME (run only once) -------------------
mycursor.execute("""
ALTER TABLE my_table 
CHANGE Adress Address VARCHAR(50)
""")
mydb.commit()

# ------------------- INSERT DATA -------------------
sql = "INSERT INTO my_table(id, Name, Age, Address) VALUES (%s, %s, %s, %s)"
val = [
    (1, "Alishba", 21, "Karachi"),
    (2, "Kinza", 20, "Lahore")
]

mycursor.executemany(sql, val)
mydb.commit()

print(mycursor.rowcount, "record inserted")
```

```python
import tkinter as tk
from tkinter import messagebox
from tkinter.ttk import Combobox
from tkinter import scrolledtext
import mysql.connector  # you forgot to import this

window = tk.Tk()
window.title("Graphical user interface")

# -------------------- Name --------------------
l1 = tk.Label(window, text="Name:", font=("Arial",15))
l1.place(x=10, y=10)
t1 = tk.Entry(window, bd=5, width=25)
t1.place(x=180, y=10)

# -------------------- Age --------------------
l2 = tk.Label(window, text="Age:", font=("Arial",15))
l2.place(x=10, y=60)
t2 = tk.Entry(window, bd=5, width=25)
t2.place(x=180, y=60)

# -------------------- Address --------------------
l3 = tk.Label(window, text="Address:", font=("Arial",15))
l3.place(x=10, y=110)
t3 = tk.Entry(window, bd=5, width=25)
t3.place(x=180, y=110)

# -------------------- Country (Drop-down) --------------------
l4 = tk.Label(window, text="Country:", font=("Arial",15))
l4.place(x=10, y=160)

c1 = Combobox(window, width=22)
c1['values'] = ("Pakistan", "India", "Nepal", "USA", "Bangladesh", "Australia")
c1.current(0)
c1.place(x=180, y=160)

# -------------------- Hobbies (Checkboxes) --------------------
l5 = tk.Label(window, text="Hobbies:", font=("Arial",15))
l5.place(x=10, y=210)

var1 = tk.IntVar()
var2 = tk.IntVar()
var3 = tk.IntVar()
var4 = tk.IntVar()

cb1 = tk.Checkbutton(window, text="Reading", font=('Arial',10), variable=var1)
cb1.place(x=180, y=210)

cb2 = tk.Checkbutton(window, text="Sports", font=('Arial',10), variable=var2)
cb2.place(x=280, y=210)

cb3 = tk.Checkbutton(window, text="Music", font=('Arial',10), variable=var3)
cb3.place(x=180, y=240)

cb4 = tk.Checkbutton(window, text="Traveling", font=('Arial',10), variable=var4)
cb4.place(x=280, y=240)

# -------------------- Sex (Radio Buttons) --------------------
l6 = tk.Label(window, text="Sex:", font=("Arial",15))
l6.place(x=10, y=290)

gender = tk.IntVar()
r1 = tk.Radiobutton(window, text="Male", variable=gender, value=1)
r1.place(x=180, y=290)

r2 = tk.Radiobutton(window, text="Female", variable=gender, value=2)
r2.place(x=260, y=290)

r3 = tk.Radiobutton(window, text="Other", variable=gender, value=3)
r3.place(x=350, y=290)

# -------------------- Comment (Scrolled Text) --------------------
l7 = tk.Label(window, text="Comment:", font=("Arial",15))
l7.place(x=10, y=340)

txt = scrolledtext.ScrolledText(window, width=23, height=3)
txt.place(x=180, y=340)
txt.insert(tk.INSERT, "Write some comments...")

# -------------------- Percentage --------------------
l8 = tk.Label(window, text="Percentage:", font=("Arial",15))
l8.place(x=10, y=410)

spin = tk.Spinbox(window, from_=0, to=100, width=5)
spin.place(x=180, y=410)

# -------------------- Submit Button --------------------
def btn_clicked():
    mydb = mysql.connector.connect(
         host="localhost",
         user="root",
         password="",
         database="python_database"
    )
    mycursor = mydb.cursor()
    sql = "INSERT INTO my_table(Name, Age, Address) VALUES (%s, %s, %s, %s)"

    # to get check button value
    if var1.get()==1 and var2.get()==1:
        s1 = "Reading and Sports"
    elif var1.get()==1:
        s1 = "Reading"
    elif var2.get()==1:
        s1 = "Sports"
    elif var3.get()==1 and var4.get()==1:
        s1 = "Music and Traveling"
    elif var3.get()==1:
        s1 = "Music"
    elif var4.get()==1:
        s1 = "Traveling"
    else:
        s1 = ""

    # to get radio button value
    if gender.get() == 1:
        s2 = "Male"
    elif gender.get() == 2:
        s2 = "Female"
    elif gender.get() == 3:
        s2 = "Other"
    else:
        s2 = ""

    val = (t1.get(), t2.get(), t3.get(), c1.get(), s1, s2, txt.get("1.0", tk.END), spin.get())
    mycursor.execute(sql, val)
    mydb.commit()
    messagebox.showinfo("Message box", "Your data is successfully inserted...")
    
b1 = tk.Button(window, text="Submit", font=("Arial",12), command=btn_clicked)
b1.place(x=160, y=460)

window.geometry("450x520")
window.mainloop()
```
