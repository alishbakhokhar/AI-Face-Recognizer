```python
import tkinter as tk
from tkinter import messagebox
import mysql.connector

# -------------------- Functions -------------------- #
def insert_user():
    if t1.get() == "" or t2.get() == "" or t3.get() == "":
        messagebox.showerror("Error", "Please fill all fields")
        return
    try:
        mydb = mysql.connector.connect(
            host="localhost",
            user="root",
            password="",  # your MySQL password
            database="authorized_user"
        )
        mycursor = mydb.cursor()
        sql = "INSERT INTO my_table(Name, Age, Address) VALUES (%s, %s, %s)"
        val = (t1.get(), int(t2.get()), t3.get())
        mycursor.execute(sql, val)
        mydb.commit()
        messagebox.showinfo("Success", "Data inserted successfully!")
        t1.delete(0, tk.END)
        t2.delete(0, tk.END)
        t3.delete(0, tk.END)
    except mysql.connector.Error as e:
        messagebox.showerror("Database Error", f"{e}")
    except ValueError:
        messagebox.showerror("Error", "Age must be a number")

# -------------------- GUI -------------------- #
window = tk.Tk()
window.title("User Entry Form")

# Name
l1 = tk.Label(window, text="Name:", font=("Arial", 15))
l1.place(x=10, y=10)
t1 = tk.Entry(window, bd=5, width=25)
t1.place(x=120, y=10)

# Age
l2 = tk.Label(window, text="Age:", font=("Arial", 15))
l2.place(x=10, y=60)
t2 = tk.Entry(window, bd=5, width=25)
t2.place(x=120, y=60)

# Address
l3 = tk.Label(window, text="Address:", font=("Arial", 15))
l3.place(x=10, y=110)
t3 = tk.Entry(window, bd=5, width=25)
t3.place(x=120, y=110)

# Submit Button
b1 = tk.Button(window, text="Submit", font=("Arial", 12), command=insert_user)
b1.place(x=150, y=160)

window.geometry("400x220")
window.mainloop()
```
